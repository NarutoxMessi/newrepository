```python
#Student Grade Calculator
Name=input("enter student name")
maths=int(input("enter student marks(0-100):"))
science=int(input("enter student marks(0-100):"))
computer=int(input("enter student marks(0-100):"))
total = maths+science+computer
percentage=total/3
print("student name:", Name)
print("student mark:", maths, science, computer)
print("student percentage",percentage)
if percentage >=75:
  print("grade A")
elif percentage >=60:
  print("grade B")
elif percentage>=40:
  print("grade c")
else:
  print("grade f")
```
