```python
#calculator using if elif else statement
num1 = int(input("enter 1st number"))
num2 = int(input("enter 2nd number"))
op = input("select a operator +,-,*,/:")
if op == "+":
  result = num1+num2
  print(result)
elif op == "-":
  result = num1-num2
  print(result)
elif op == "*":
  result = num1*num2
  print(result)
elif op == "/":
  if num2!=0:
    result = num1/num2
    print(result)
  else:
    print("divisible by 0 is infinity")
else:
  print("select correct operator in the list")
```
